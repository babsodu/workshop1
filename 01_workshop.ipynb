## Find largest number in a list

```python
def find_largest(numbers):
    """
    This function takes a list of numbers and returns the largest number.
    
    Parameters:
    numbers (list): List of numbers
    
    Returns:
    int/float: The largest number in the list
    """
    if not numbers: 
        return None
    
    largest = numbers[0] 
    
    for num in numbers:
        if num > largest:
            largest = num
    
    return largest
```

```python
list1 = [3, 7, 2, 9, 5]
result1 = find_largest(list1)
print(f"find_largest({list1}) = {result1}")

list2 = [-5, -2, -1, -9]
result2 = find_largest(list2)
print(f"find_largest({list2}) = {result2}")
```

## Count vowels in a string

```python
def count_vowels(text):
    """
    This function takes a string and returns the number of vowels.
    
    Parameters:
    text (str): Input string
    
    Returns:
    int: Number of vowels in the string
    """
    vowels = "aeiouAEIOU"
    count = 0
    
    for char in text:
        if char in vowels:
            count += 1
    
    return count
```

```python
string1 = "Python Programming"
result1 = count_vowels(string1)
print(f"count_vowels('{string1}') = {result1}")


string2 = "AEIOUaeiou"
result2 = count_vowels(string2)
print(f"count_vowels('{string2}') = {result2}")
```

# Exercise 6: Password Strength Checker

```python
special_chars = "!@#$%^&*()-_=+[]{}|;:,.<>?/`~\"'\\"

def check_password_strength(password):
    """
    Check password strength and return score and feedback
    """
    score = 0
    feedback = []
 
    length = len(password)
    if length >= 8:
        score += 1
        feedback.append("Meets minimum length requirement (8+ characters)")
    else:
        feedback.append("Too short (minimum 8 characters required)")
    
    has_upper = any(char.isupper() for char in password)
    if has_upper:
        score += 1
        upper_count = sum(1 for char in password if char.isupper())
        feedback.append(f"Contains uppercase letters ({upper_count} found)")
    else:
        feedback.append("No uppercase letters found")
    
    has_lower = any(char.islower() for char in password)
    if has_lower:
        score += 1
        lower_count = sum(1 for char in password if char.islower())
        feedback.append(f"Contains lowercase letters ({lower_count} found)")
    else:
        feedback.append("No lowercase letters found")
    
    has_special = any(char in special_chars for char in password)
    if has_special:
        score += 1
        special_count = sum(1 for char in password if char in special_chars)
        feedback.append(f"Contains special characters ({special_count} found)")
    else:
        feedback.append("No special characters found")
    
    has_digit = any(char.isdigit() for char in password)
    if has_digit:
        score += 1
        digit_count = sum(1 for char in password if char.isdigit())
        feedback.append(f"Contains digits ({digit_count} found)")
    else:
        feedback.append("No digits found (recommended for added security)")
    
    if score >= 5:
        strength = "Your Password is VERY STRONG"
    elif score == 4:
        strength = "Your Password is STRONG"
    elif score == 3:
        strength = "Your Password is MODERATE"
    elif score == 2:
        strength = "Your Password is WEAK"
    else:
        strength = "Your Password is VERY WEAK"
    
    return score, strength, feedback
```

```python
pwd = input("Enter password to check: ")

if pwd:
    score, strength, feedback = check_password_strength(pwd)
    
    print(f"\nRESULT: {strength}")
    print(f"   Score: {score}/5")
    
    print("\nFeedbacks:")
    for item in feedback:
        print(f"   {item}")
    
else:
    print("Error: Password cannot be empty!")
```

# Exercise 7: Caesar Cipher

```python
def caesar_cipher(text, shift, mode='encrypt'):
    """
    This function handles 
    - Uppercase letters (A-Z)
    - Lowercase letters (a-z)
    - Preserves non-alphabetic characters
    - Handles both encryption and decryption
    """
    result = ""
    
    for char in text:
        if char.isupper():
            ascii_val = ord(char)
            if mode == 'encrypt':
                shifted = (ascii_val - 65 + shift) % 26 + 65
            else:  # decrypt
                shifted = (ascii_val - 65 - shift) % 26 + 65
            result += chr(shifted)
            
        elif char.islower():
            ascii_val = ord(char)
            if mode == 'encrypt':
                shifted = (ascii_val - 97 + shift) % 26 + 97
            else:  # decrypt
                shifted = (ascii_val - 97 - shift) % 26 + 97
            result += chr(shifted)
            
        else:
            result += char
    
    return result
```

```python
message = "HELLOiy"
key = 3

print(f"Original message: {message}")
print(f"Key: {key}")

encrypted = caesar_cipher(message, key, 'encrypt')
print(f"Encrypted: {encrypted}")

decrypted = caesar_cipher(encrypted, key, 'decrypt')
print(f"Decrypted: {decrypted}")

```

# Exercise 8: Hangman Game

```python
import random
import string


class HangmanGame:
    
    def __init__(self, word_list=None, max_guesses=6):
        if word_list is None:
            self.word_list = [
                "python", "java", "javascript", "programming", "computer",
                "algorithm", "function", "variable", "dictionary", "string",
                "integer", "boolean", "float", "list", "tuple", "set"
            ]
        else:
            self.word_list = word_list
        
        self.max_guesses = max_guesses
        self.secret_word = ""
        self.guessed_letters = set()
        self.incorrect_guesses = 0
        self.game_won = False
        self.game_over = False
    
    def start_game(self):
        """Start a new game"""
        self.secret_word = random.choice(self.word_list).lower()
        self.guessed_letters = set()
        self.incorrect_guesses = 0
        self.game_won = False
        self.game_over = False
        
        print("NEW HANGMAN GAME STARTED!")
        print(f"Word length: {len(self.secret_word)} letters")
        print(f"Maximum incorrect guesses: {self.max_guesses}")
        print(f"Category: Programming Languages & Terms")
    
    def display_status(self):
        """Display current game status"""

        displayed_word = self.get_displayed_word()
        print(f"\nWord: {displayed_word}")
        
        guessed_str = ", ".join(sorted(self.guessed_letters)) if self.guessed_letters else "None"
        print(f"Guessed letters: {guessed_str}")
        
        print(f"Incorrect guesses: {self.incorrect_guesses}/{self.max_guesses}")
        remaining = self.max_guesses - self.incorrect_guesses
        print(f"Guesses remaining: {remaining}")
        
        # Show hint if struggling
        if remaining <= 2 and self.incorrect_guesses > 0:
            self.give_hint()
   
    def get_displayed_word(self):
        """Get the word with underscores for unguessed letters"""
        result = []
        for letter in self.secret_word:
            if letter in self.guessed_letters:
                result.append(letter)
            else:
                result.append("_")
        return " ".join(result)
    
    def process_guess(self, guess):
        """Process a player's guess"""
        guess = guess.lower()
        
        if len(guess) != 1:
            return "Please enter only one letter!"
        
        if not guess.isalpha():
            return "Please enter a letter (a-z)!"
        
        if guess in self.guessed_letters:
            return f"You already guessed '{guess}'!"
        
        self.guessed_letters.add(guess)
        
        if guess in self.secret_word:
            # Check if player has won
            if all(letter in self.guessed_letters for letter in self.secret_word):
                self.game_won = True
                self.game_over = True
                return "win"
            return "correct"
        else:
            self.incorrect_guesses += 1
            if self.incorrect_guesses >= self.max_guesses:
                self.game_over = True
                return "lose"
            return "incorrect"
    
    def give_hint(self):
        """Give a hint if player is struggling"""
        unguessed = [letter for letter in self.secret_word if letter not in self.guessed_letters]
        if unguessed:
            # Give vowel hint if available
            vowels = [l for l in unguessed if l in 'aeiou']
            if vowels:
                print(f"Hint: Try guessing a vowel like '{vowels[0]}'")
            else:
                print(f"Hint: The word contains '{unguessed[0]}'")
    
    def play(self):
        """Play a complete game"""
        self.start_game()
        
        while not self.game_over:
            self.display_status()
            
            guess = input("\nGuess a letter (or type 'quit' to exit): ").lower()
            
            if guess == 'quit':
                print(f"\nGame ended. The word was: '{self.secret_word}'")
                return
            
            # Process the guess
            result = self.process_guess(guess)
            
            if result == "win":
                print(f"\n{'='*50}")
                print(f"CONGRATULATIONS! YOU WIN!")
                print(f"   The word was: '{self.secret_word}'")
                print(f"   You had {self.max_guesses - self.incorrect_guesses} guesses remaining")
            elif result == "lose":
                print(f"\n{'='*50}")
                print(f"GAME OVER! YOU LOSE!")
                print(f"   The word was: '{self.secret_word}'")
                print(f"   Your guesses: {', '.join(sorted(self.guessed_letters))}")
            elif result == "correct":
                print(f"Good guess! '{guess}' is in the word.")
            elif result == "incorrect":
                print(f"Sorry, '{guess}' is not in the word.")
            else:
                print(f"{result}")
```

```python
game = HangmanGame()
game.play()
```
